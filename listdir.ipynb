```python
import os

def list_files_recursively(dir_path):
    # dir_path 내부의 모든 항목(파일, 디렉토리) 목록
    items = os.listdir(dir_path)
    for item in items:
        # 전체 경로
        full_path = os.path.join(dir_path, item)
        # 만약 디렉토리면 재귀 호출로 하위 파일까지 탐색
        if os.path.isdir(full_path):
            list_files_recursively(full_path)
        else:
            # 파일이면 경로 출력
            print(full_path)

if __name__ == "__main__":
    # 현재 작업 디렉토리를 기준으로 재귀 탐색
    current_dir = os.getcwd()
    list_files_recursively(current_dir)
```
